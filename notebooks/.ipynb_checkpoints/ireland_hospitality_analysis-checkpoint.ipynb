```python
from sqlalchemy import create_engine, text
import pandas as pd
import matplotlib.pyplot as plt

engine = create_engine('postgresql://postgres:203040@localhost:5432/postgres')

conn = engine.connect().execution_options(no_parameters=True)

query = """
SELECT year, 
       SUM(active_enterprises)::INT AS total_enterprises,
       SUM(persons_engaged)::INT    AS total_persons
FROM hospitality.hospitality_county_year
WHERE activity ILIKE '%Accommodation and Food Service%'
GROUP BY year
ORDER BY year;
"""

df = pd.read_sql(text(query), engine)
print(df)
```

```python
import matplotlib.pyplot as plt

fig, ax1 = plt.subplots(figsize=(10, 6))

ax1.bar(df['year'], df['total_enterprises'], color='steelblue', alpha=0.7, label='Enterprises')
ax1.set_xlabel('Year')
ax1.set_ylabel('Active Enterprises', color='steelblue')
ax1.tick_params(axis='y', labelcolor='steelblue')

ax2 = ax1.twinx()
ax2.plot(df['year'], df['total_persons'], color='darkred', marker='o', linewidth=2.5, label='Persons Engaged')
ax2.set_ylabel('Persons Engaged', color='darkred')
ax2.tick_params(axis='y', labelcolor='darkred')

plt.title('Ireland Hospitality Sector — Enterprises & Employment 2019–2023', fontsize=13, fontweight='bold')
fig.legend(loc='upper left', bbox_to_anchor=(0.1, 0.95))
plt.tight_layout()
plt.savefig('chart_01_national_overview.png', dpi=150, bbox_inches='tight')
plt.show()
print("Chart saved!")
```

## 1. National Overview — Enterprises & Employment 2019–2023

The chart below shows the full picture of Ireland's hospitality sector over five years.
The bars represent the number of active businesses, while the line tracks total employment.

**What the data shows:**
- 2020: COVID-19 caused the sector to lose **63,868 jobs** in a single year — a drop of 28.9%
- 2021: Slow recovery began as restrictions eased
- 2022: Employment surpassed pre-pandemic levels for the first time
- 2023: New historic high — **242,791 persons engaged**, up 9.8% vs 2019

```python
query2 = """
SELECT 
    a.county,
    a.persons_engaged AS persons_2019,
    b.persons_engaged AS persons_2020,
    ROUND(100.0 * (b.persons_engaged - a.persons_engaged) 
          / a.persons_engaged, 1) AS pct_drop
FROM hospitality.hospitality_county_year a
JOIN hospitality.hospitality_county_year b 
    ON a.county = b.county
WHERE a.year = 2019 
  AND b.year = 2020
  AND a.activity ILIKE '%Accommodation and Food Service%'
  AND b.activity ILIKE '%Accommodation and Food Service%'
  AND a.county != 'Ireland'
ORDER BY pct_drop ASC;
"""

df2 = pd.read_sql(query2, conn)

plt.figure(figsize=(12, 7))
colors = ['darkred' if x < -30 else 'tomato' if x < -25 else 'steelblue' for x in df2['pct_drop']]
bars = plt.barh(df2['county'], df2['pct_drop'], color=colors)
plt.axvline(x=df2['pct_drop'].mean(), color='black', linestyle='--', linewidth=1.5, label=f'Average: {df2["pct_drop"].mean():.1f}%')
plt.xlabel('Employment Change (%)')
plt.title('COVID-19 Impact on Hospitality Employment by County — 2019 vs 2020', fontsize=13, fontweight='bold')
plt.legend()
plt.tight_layout()
plt.savefig('chart_02_covid_impact_by_county.png', dpi=150, bbox_inches='tight')
plt.show()
print("Chart saved!")
```

## 2. COVID-19 Impact by County — Employment Drop in 2020

This chart shows which counties were hit hardest by COVID-19 restrictions in 2020,
measured by the percentage drop in persons engaged compared to 2019.

**What the data shows:**
- Every single county in Ireland lost hospitality jobs in 2020
- The national average drop was -28.9%
- Counties heavily dependent on international tourism suffered the most
- The dashed line shows the national average — counties to the left were hit harder than average

```python
query3 = """
SELECT 
    a.county,
    ROUND(100.0 * b.active_enterprises / a.active_enterprises, 1) AS recovery_index,
    CASE
        WHEN b.active_enterprises >= a.active_enterprises * 1.10 THEN 'Strong growth (+10%)'
        WHEN b.active_enterprises >= a.active_enterprises * 1.05 THEN 'Good growth (+5%)'
        WHEN b.active_enterprises >= a.active_enterprises        THEN 'Fully recovered'
        WHEN b.active_enterprises >= a.active_enterprises * 0.95 THEN 'Nearly recovered'
        ELSE 'Still below 2019'
    END AS recovery_status
FROM hospitality.hospitality_county_year a
JOIN hospitality.hospitality_county_year b 
    ON a.county = b.county
WHERE a.year = 2019 
  AND b.year = 2023
  AND a.activity ILIKE '%Accommodation and Food Service%'
  AND b.activity ILIKE '%Accommodation and Food Service%'
ORDER BY recovery_index DESC;
"""

df3 = pd.read_sql(query3, conn)

colors_map = {
    'Strong growth (+10%)': '#1a7a4a',
    'Good growth (+5%)':    '#5cb85c',
    'Fully recovered':      '#5bc0de',
    'Nearly recovered':     '#f0ad4e',
    'Still below 2019':     '#d9534f'
}
colors3 = [colors_map[s] for s in df3['recovery_status']]

plt.figure(figsize=(12, 7))
bars = plt.barh(df3['county'], df3['recovery_index'], color=colors3)
plt.axvline(x=100, color='black', linestyle='--', linewidth=1.5, label='2019 baseline (100)')

from matplotlib.patches import Patch
legend_elements = [Patch(facecolor=v, label=k) for k, v in colors_map.items()]
plt.legend(handles=legend_elements, loc='lower right', fontsize=9)

plt.xlabel('Recovery Index (2019 = 100)')
plt.title('Hospitality Sector Recovery by County — 2023 vs 2019 Baseline', fontsize=13, fontweight='bold')
plt.tight_layout()
```

## 3. Recovery Index by County — 2023 vs 2019 Baseline

This chart measures how well each county recovered by 2023 compared to pre-pandemic 2019 levels.
A recovery index above 100 means the county has surpassed its 2019 position.

**What the data shows:**
- The majority of Irish counties fully recovered or exceeded 2019 levels by 2023
- Green counties show strong growth — these are the resilience stories
- Counties still in red represent areas where the hospitality sector has not yet returned to pre-COVID levels
- This highlights the uneven nature of recovery across Ireland

```python
query4 = """
SELECT 
    year,
    CASE 
        WHEN activity = 'Accommodation (I55)' THEN 'Accommodation'
        WHEN activity = 'Food and Beverage Service Activities (I56)' THEN 'Food & Beverage'
    END AS sub_sector,
    SUM(active_enterprises)::INT AS enterprises,
    SUM(persons_engaged)::INT    AS persons
FROM hospitality.hospitality_county_year
WHERE activity IN (
    'Accommodation (I55)',
    'Food and Beverage Service Activities (I56)'
)
GROUP BY year, activity
ORDER BY sub_sector, year;
"""

df4 = pd.read_sql(query4, conn)

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

for ax, metric, title in zip(
    axes,
    ['enterprises', 'persons'],
    ['Active Enterprises', 'Persons Engaged']
):
    for sub, color in [('Accommodation', 'steelblue'), ('Food & Beverage', 'darkorange')]:
        data = df4[df4['sub_sector'] == sub]
        ax.plot(data['year'], data[metric], marker='o', linewidth=2.5,
                color=color, label=sub)
    ax.set_title(f'{title} by Sub-Sector', fontweight='bold')
    ax.set_xlabel('Year')
    ax.set_ylabel(title)
    ax.legend()
    ax.grid(axis='y', alpha=0.3)

fig.suptitle('Accommodation vs Food & Beverage — Recovery Comparison 2019-2023',
             fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig('chart_04_subsector_comparison.png', dpi=150, bbox_inches='tight')
plt.show()
print("Chart saved!")
```

## 4. Accommodation vs Food & Beverage — Recovery Comparison

This chart compares the two main sub-sectors of Irish hospitality:
Accommodation (hotels, B&Bs) vs Food & Beverage (restaurants, bars, cafés).

**What the data shows:**
- Accommodation was hit harder in 2020 — hotels were fully closed during lockdowns
- Food & Beverage showed more resilience — takeaway and delivery kept some businesses operating
- By 2023 both sub-sectors surpassed 2019 levels
- Food & Beverage employs significantly more people than Accommodation across all years

```python
query5 = """
SELECT 
    year,
    SUM(CASE WHEN statistic = 'Enterprise Births' 
        AND value_text != '' AND value_text IS NOT NULL 
        THEN value_text::NUMERIC ELSE 0 END) AS new_businesses,
    SUM(CASE WHEN statistic = 'Final Enterprise Deaths' 
        AND value_text != '' AND value_text IS NOT NULL 
        THEN value_text::NUMERIC ELSE 0 END) AS closed_businesses
FROM hospitality.enterprise_size_data
WHERE activity ILIKE '%Accommodation and Food Service%'
  AND employment_size = 'All persons engaged size classes'
GROUP BY year
HAVING SUM(CASE WHEN statistic = 'Enterprise Births' 
    AND value_text != '' THEN value_text::NUMERIC ELSE 0 END) > 0
ORDER BY year;
"""

df5 = pd.read_sql(query5, conn)
df5['ratio'] = (df5['new_businesses'] / df5['closed_businesses']).round(2)

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

colors = ['#d9534f' if x < 1 else '#5cb85c' for x in df5['ratio']]
axes[0].bar(df5['year'], df5['ratio'], color=colors)
axes[0].axhline(y=1, color='black', linestyle='--', linewidth=1.5, label='Break-even (ratio = 1)')
axes[0].set_title('Birth-to-Death Ratio by Year', fontweight='bold')
axes[0].set_xlabel('Year')
axes[0].set_ylabel('Ratio')
axes[0].legend()
for i, v in enumerate(df5['ratio']):
    axes[0].text(df5['year'].iloc[i], v + 0.01, str(v), ha='center', fontsize=10)

axes[1].bar(df5['year'], df5['new_businesses'], color='steelblue', label='New businesses')
axes[1].bar(df5['year'], -df5['closed_businesses'], color='tomato', label='Closed businesses')
axes[1].axhline(y=0, color='black', linewidth=0.8)
axes[1].set_title('New vs Closed Businesses per Year', fontweight='bold')
axes[1].set_xlabel('Year')
axes[1].set_ylabel('Number of Businesses')
axes[1].legend()

fig.suptitle('Ireland Hospitality — Business Birth & Death Analysis 2019-2023',
             fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig('chart_05_birth_death_ratio.png', dpi=150, bbox_inches='tight')
plt.show()
print("Chart saved!")
```

## 5. Business Birth-to-Death Ratio — Sector Health 2019–2023

This chart reveals whether more businesses were opening or closing each year.
A ratio above 1.0 means more openings than closures — a healthy, growing sector.
A ratio below 1.0 means more closures than openings — a contracting sector.

**What the data shows:**
- 2019: ratio of 0.90 — the sector was already contracting before COVID
- 2020: ratio of 0.89 — COVID made things worse but the drop was not as dramatic as expected
- 2021: ratio of 0.54 — the worst year on record, only 821 new businesses opened against 1,513 closures
- This data suggests the sector faced structural challenges even before the pandemic

```python
query_findings = """
SELECT 
    year,
    SUM(active_enterprises)::INT AS enterprises,
    SUM(persons_engaged)::INT AS persons,
    ROUND(100.0 * SUM(active_enterprises) / 
        FIRST_VALUE(SUM(active_enterprises)) OVER (ORDER BY year), 1) AS index_vs_2019
FROM hospitality.hospitality_county_year
WHERE activity ILIKE '%Accommodation and Food Service%'
GROUP BY year
ORDER BY year;
"""
df_findings = pd.read_sql(query_findings, conn)
print(df_findings.to_string())
```

```python
query_counties = """
SELECT 
    a.county,
    a.persons_engaged::INT AS persons_2019,
    b.persons_engaged::INT AS persons_2023,
    (b.persons_engaged - a.persons_engaged)::INT AS jobs_gained,
    ROUND(100.0 * (b.persons_engaged - a.persons_engaged) / a.persons_engaged, 1) AS pct_change
FROM hospitality.hospitality_county_year a
JOIN hospitality.hospitality_county_year b ON a.county = b.county
WHERE a.year = 2019 AND b.year = 2023
  AND a.activity ILIKE '%Accommodation and Food Service%'
  AND b.activity ILIKE '%Accommodation and Food Service%'
ORDER BY pct_change DESC;
"""
df_counties = pd.read_sql(query_counties, conn)
print(df_counties.to_string())
```
